Стохастический градиентный спуск — итерационный метод для оптимизации целевой функции с подходящими свойствами гладкости. Его можно расценивать как стохастическую аппроксимацию оптимизации методом градиентного спуска, поскольку он заменяет реальный градиент, вычисленный из полного набора данных, оценкой, вычисленной из случайно выбранного подмножества данных.

```python
import random
import math
```

```python
def predict(weights, bias, X):
    if isinstance(X[0], (int, float)):  # один пример
        return sum(w * x for w, x in zip(weights, X)) + bias
    else:  # несколько примеров
        return [sum(w * x for w, x in zip(weights, row)) + bias for row in X]

def sgd_step(weights, bias, x_i, y_i, learning_rate):
    # предсказание
    prediction = predict(weights, bias, x_i)
    error = prediction - y_i
    
    # обновляем веса
    new_weights = []
    for j in range(len(weights)):
        grad_w = error * x_i[j]
        new_weights.append(weights[j] - learning_rate * grad_w)
    
    # обновляем bias
    new_bias = bias - learning_rate * error
    
    return new_weights, new_bias

def sgd_fit(X, y, learning_rate=0.01, epochs=100, verbose=True):# обучение SGD
    n_features = len(X[0])
    weights = [0.0] * n_features
    bias = 0.0
    loss_history = []
    
    for epoch in range(epochs):
        indices = list(range(len(X)))
        random.shuffle(indices) # перемешиваем данные
        
        epoch_loss = 0
        
        for idx in indices: # проходим по каждому примеру
            x_i = X[idx]
            y_i = y[idx]
            
            weights, bias = sgd_step(weights, bias, x_i, y_i, learning_rate)
            
            pred = predict(weights, bias, x_i) # считаем ошибку
            epoch_loss += (pred - y_i) ** 2
        
        avg_loss = epoch_loss / len(X)
        loss_history.append(avg_loss)
        
        if verbose and epoch % 10 == 0:
            print(f"Epoch {epoch:3d}, Loss: {avg_loss:.6f}")
    
    return weights, bias, loss_history
```

```python
# пример использования
X = [[x] for x in range(1, 101)]
y = [3*x[0] + 2 + random.uniform(-2, 2) for x in X]

weights, bias, _ = sgd_fit(X, y, learning_rate=0.0001, epochs=100)
print(f"Результат: вес = {weights[0]}, bias = {bias}")
```
