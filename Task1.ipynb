```python
def count_backtracking():
    all_rows = [(0,1), (0,2), (0,3), (1,2), (1,3), (2,3)]
    def backtrack(row_idx, col_counts):
        if row_idx == 4:
            return 1 if col_counts == [2, 2, 2, 2] else 0
        
        #если в каком-то столбце уже больше 2 или не хватит строк до 2
        remaining_rows = 4 - row_idx
        for c in range(4):
            if col_counts[c] > 2:
                return 0
            if col_counts[c] + remaining_rows < 2:
                return 0
        
        total = 0
        for row in all_rows:
            new_counts = col_counts.copy()
            for c in row:
                new_counts[c] += 1
            total += backtrack(row_idx + 1, new_counts)
        return total
    
    return backtrack(0, [0, 0, 0, 0])

print(f"Число матриц (backtracking): {count_backtracking()}")
```

```python
#задача сводится к матрице 4×4
#где каждый блок либо горизонтальные домино, либо вертикальные
#горизонтальные = 1 в матрице, вертикальные = 0
def count_domino_tilings():
    from itertools import combinations
    all_rows = list(combinations(range(4), 2))  #6 вариантов расположения горизонтальных блоков в строке блоков
    count = 0
    solutions = []
    for r1 in all_rows:
        col_counts = [0] * 4
        for c in r1: 
            col_counts[c] += 1    
        for r2 in all_rows:
            cc2 = col_counts.copy()
            for c in r2: 
                cc2[c] += 1
            if any(x > 2 for x in cc2):
                continue    
            for r3 in all_rows:
                cc3 = cc2.copy()
                for c in r3: 
                    cc3[c] += 1
                if any(x > 2 for x in cc3):
                    continue    
                for r4 in all_rows:
                    cc4 = cc3.copy()
                    for c in r4: 
                        cc4[c] += 1
                    if cc4 == [2, 2, 2, 2]:
                        count += 1
                        matrix = [[0]*4 for _ in range(4)]
                        for c in r1: matrix[0][c] = 1
                        for c in r2: matrix[1][c] = 1
                        for c in r3: matrix[2][c] = 1
                        for c in r4: matrix[3][c] = 1
                        solutions.append(matrix)
    return count, solutions

count, solutions = count_domino_tilings()
print("Число способов разрезания:", count)
```
